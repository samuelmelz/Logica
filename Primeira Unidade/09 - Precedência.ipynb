# Precedência

Assim como a matemática, a lógica tem operações que, se feitas na ordem errada, geram resultados diferentes. Para evitar esse tipo de coisa, existe a precedência. Abaixo há a lista da maior para a menor precedência.

1 - ( ) - Parênteses

2 - Não - Negação

3 - E - Conjunção

4 - Ou - Disjunção

5 - Se... então - Implicação

6 - Se, e somente se - Bi-implicação

#### Exemplo 1: 

* ((¬(¬p)) -> q)

Na expressão acima, como há parênteses, devemos resolvê-los primeiro, sempre do mais interno ao mais externo. Para fins didáticos, considere p = V e q = F. Substituindo esses valores na expressão, temos:

* ((¬(¬V)) -> F)

Os parênteses mais internos são **(¬V)**. Portanto, realizamos essa negação, obtendo o valor F.

* ((¬F) -> F)

Agora os parênteses mais internos são **(¬F)**. Realizando essa negação, temos:

* (V->F)

Agora chegamos ao caso simples, que já conhecemos

* F

#### Exemplo 2:

* (p v q ^ r) ^ (p -> r)

Aqui temos dois parênteses que tem a mesma precedência, já que nenhum é mais interno que outro. Considerando p = V, q = F e r = V, temos:

* (V v F ^ V) ^ (V -> V)

No primeiro par de parênteses, temos duas operações distintas, porém, pela ordem, sabemos que devemos realizar primeiro a conjunção. 

* (V v F) ^ (V -> V)

Uma vez que realizamos a conjunção, nos sobra um caso simples.

* V ^ (V -> V)

Não podemos agora realizar essa conjunção, pois ainda há parênteses esperando por sua solução, **(V -> V)**. Este é um caso simples de implicação, temos:

* V ^ V

Agora finalmente podemos realizar a conjunção.

* V
